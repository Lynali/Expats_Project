```python
import numpy as np
import matplotlib.pyplot as plt

mean = 717203.9333
std_dev = 527507.7798
sample_size = 30

data = np.random.normal(loc=mean, scale=std_dev, size=sample_size)

plt.hist(data, bins=6, density=True, alpha=0.6, color='b', edgecolor='black')

plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')

x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)
pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

plt.plot(x, pdf, 'r', linewidth=2)

plt.title('Bell Curve (Normal Distribution) with Sample Size 30 (European Countries Net Migration')
plt.xlabel('Value')
plt.ylabel('Density')

plt.show()
```
